```python
import holoviews as hv
import numpy as np
import panel as pn

hv.extension('bokeh')
```

```python
points = hv.Points([(0.0, 0.0), (1.0, 1.0)]).opts(size=10)

annotator = hv.annotate.instance()
layout = annotator(points, annotations=['Label'], name="Point Annotations")

print(layout)
```
